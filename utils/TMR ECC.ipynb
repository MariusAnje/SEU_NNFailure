```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def find_k(x):
    i = 0
    ta = []
    for num in x:
        while True:
            if 2**i - 1 >= i+num:
                break
            i += 1
        ta += [i]
    return np.array(ta)
```

```python
x = np.arange(1,100)
B = x**2 * 32
TM = (3. * x)/x
EM = (x + find_k(x))/x
plt.plot(x,TM)
plt.plot(x,EM)
plt.ylabel('relative memory usage')
plt.xlabel('number of bits protected')
label = ['TMR', 'ECC']
plt.legend(label, loc = 0, ncol = 1, edgecolor = None)
plt.savefig('TVEM')
```

```python
x = np.arange(1,5*5*32)
TL = (3. * x)
EL = find_k(x)*(np.ceil((x+find_k(x))/2.)-1)+find_k(x)*x
plt.plot(x,TL)
plt.plot(x,EL)
plt.ylabel('relative logic area usage')
plt.xlabel('number of bits protected')
label = ['TMR', 'ECC']
plt.legend(label, loc = 0, ncol = 1, edgecolor = None)
#plt.savefig('TVEArea')
```
