```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def find_k(x):
    i = 0
    ta = []
    for num in x:
        while True:
            if 2**i - 1 >= i+num:
                break
            i += 1
        ta += [i]
    return np.array(ta)
```

```python
a = plt.subplot(1,2,1)
x = np.arange(1,3*3*32)
B = x**2 * 32
TM = (2. * x)
EM = (find_k(x))
#plt.plot(x,TM)
a.plot(x,EM/TM)
a.set_gid('a')
a.set_ylabel('Relative SRAM area overhead')
a.set_xlabel('Number of bits to be protected')
#label = ['TMR', 'ECC']
#plt.legend(label, loc = 0, ncol = 1, edgecolor = None)

b = plt.subplot(1,2,2)
x = np.arange(1,3*3*32)
TL = (3. * x)
EL = find_k(x)*(np.ceil((x+find_k(x))/2.)-1)+find_k(x)*x
#plt.plot(x,TL)
b.plot(x,EL/TL)
b.set_ylabel('Relative protection logic consumption')
b.set_xlabel('Number of bits to be protected')
plt.subplots_adjust(wspace =0.5, hspace =0)
#label = ['TMR', 'ECC']
#plt.legend(label, loc = 0, ncol = 1, edgecolor = None)
#plt.savefig('TVEArea')
#plt.savefig('TVEM')
```
